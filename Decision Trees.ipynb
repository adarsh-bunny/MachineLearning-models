## Algorithm used :Decision Tree Classification Algorithm

* It is a graphical representation for getting all the possible solutions to a problem/decision based on given conditions.
* Decision Tree is a Supervised learning technique that can be used for both classification and Regression problems, but mostly   it is preferred for solving Classification problems. 
* It is a tree-structured classifier, where internal nodes represent the     features of a dataset, branches represent the decision rules and each leaf node represents the outcome.


 ## PYTHON  CODE

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
iris=pd.read_csv("datasets\\iris.csv")
iris.head()
```

```python
iris["Species"].value_counts()
```

```python
X=iris.iloc[:,:-1]
X
y=iris.iloc[:,-1]
#y
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=50)
```

```python
X_train
y_train
```

# Importing Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
dtree=DecisionTreeClassifier(criterion="gini")
```

```python
clf=dtree.fit(X_train,y_train)
```

# Visualizing DecisionTree

```python
plt.figure(figsize=(15,10))
tree.plot_tree(clf,filled=True);
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(clf,filled=True,max_depth=2);
```

# Prediction from Decision Tree

```python
y_pred=dtree.predict(X_test)
y_pred
```

```python
y_test
```

## EVALUATION OF THE MODEL METRICS

### Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
import seaborn as sn
data = confusion_matrix(y_test,y_pred)
df_cm = pd.DataFrame(data, columns=np.unique(y_test), index = np.unique(y_test))
df_cm.index.name = 'Actual'
df_cm.columns.name = 'Predicted'
plt.figure(figsize = (10,7))
sn.set(font_scale=1.4)#for label size
sn.heatmap(df_cm, cmap="Reds", annot=True,annot_kws={"size": 16})# font size
```

```python
confusion_matrix(y_test,y_pred)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, precision_score, recall_score, f1_score
y_true=y_test
cm = confusion_matrix(y_true, y_pred)
print(cm)
print("\n",accuracy_score(y_true, y_pred))
print("\n",classification_report(y_true,y_pred))
```

```python
print("Precision:",precision_score(y_true,y_pred, average='micro'))
print("Recall:",recall_score(y_true,y_pred, average="macro"))
print("f1-score:",f1_score(y_true,y_pred,average=None))
```

* **Precision**-It is the ‘Exactness’, ability of the model to return only relevant instances.
* **Recall**-It is the ‘Completeness’, ability of the model to identify all relevant instances.
* **F1 Measure**-Harmonic mean of Precision & Recall, used to indicate a balance between Precision & Recall providing each equal weightage, it ranges from 0 to 1.
* A GOOD MODEL INDICATES A HIGH VALUES IN DIAGONAL ELEMENTS AND LESS VALUES ON OFF-DIAGONALS ELEMENTS.
